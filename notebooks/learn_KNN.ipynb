```python
import pandas as pd
from sklearn import datasets
 
dataset = datasets.load_breast_cancer()
df = pd.DataFrame(dataset.data,columns=dataset.feature_names)
df['cancer'] = pd.Series(dataset.target)
df.head()
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split

X = df[dataset.feature_names]
Y = df["cancer"]

# The stratify parameter in train_test_split guarantees that the proportion of each class 
# in the target (Y) variable is preserved in both the training and testing sets

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42, stratify=Y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
knn = KNeighborsClassifier(n_neighbors=8)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
print(knn.score(X_test, y_test))
```

```python
from sklearn.model_selection import GridSearchCV
 
param_grid = {'n_neighbors':np.arange(1, 50)}
knn_cv = GridSearchCV(knn, param_grid, cv=5)
knn_cv.fit(X_train, y_train)
print(knn_cv.best_params_)
print(knn_cv.best_score_)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
Steps = [
    ('scalar', StandardScaler()),
    ('knn', KNeighborsClassifier())
]
param_grid = {'n_neighbors':np.arange(1, 50)}
pipeline = Pipeline(Steps)
knn_cv = GridSearchCV(pipeline, param_grid, cv=5)
knn_cv.fit(X_train, y_train)
print(knn_cv.best_params_)
print(knn_cv.best_score_)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
Steps = [
    ('scaler', StandardScaler()),
    ('knn', KNeighborsClassifier())
]
## Tem que colocar o knn__ na frente devido ao pipeline
param_grid = {'knn__n_neighbors':np.arange(1, 50)}
pipeline = Pipeline(Steps)
knn_cv = GridSearchCV(pipeline, param_grid, cv=5)
knn_cv.fit(X_train, y_train)
print(knn_cv.best_params_)
print(knn_cv.best_score_)
```
